## 2. The question: decision, action, cost of a wrong call

*What decision does your work improve? Who acts on it? What does a wrong recommendation cost?*

**ANSWER:**
### Decision

Determine which content pages should enter the review queue first.

### Who Takes the Action?

SEO analysts, content strategists, or content editors.

### Action

Review the highest-ranked pages first and decide whether to:

- Refresh existing content
- Expand content
- Improve metadata
- Monitor performance
- Leave the page unchanged

### Unit of Analysis

One content page.

### Output

A ranked list of content pages with:

- Priority score
- Suggested action
- Confidence level
- Explainable reason codes

### Cost of a Wrong Recommendation

A **false positive** recommends reviewing a page that does not actually require attention, wasting editor time and organizational resources.

A **false negative** is potentially more costly because an important page may continue losing search visibility or traffic without receiving timely intervention.

### Why Machine Learning Helps

Simple rules can only consider a small number of signals at one time. Search performance depends on many interacting factors including impressions, CTR, average ranking position, engagement, freshness, content age, sessions, and traffic trends.

Machine learning can learn complex relationships among these observed signals and produce a better ranked recommendation list while still supporting—not replacing—human decision making.
