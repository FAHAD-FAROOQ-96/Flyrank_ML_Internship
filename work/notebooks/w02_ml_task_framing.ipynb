## 1. My lane as an ML task (type)

*Classification, clustering, ranking, or scoring — which one, and why?*

**ANSWER:**
### Selected Lane

**Lane 2 – Refresh / Content Opportunity Scoring**

### ML Task Type

**Ranking / Scoring**

The objective is not simply to classify pages as "good" or "bad." Instead, the system should assign every content page a priority score and rank pages according to how urgently they should be reviewed.

This ranking supports content teams by helping them decide which pages deserve attention first. Since editorial resources are limited, producing an ordered priority list is more useful than making independent yes/no predictions for each page.

## 2. Target or proxy

*What would you predict? Where does that label come from — observed outcome or a defined rule?*

**ANSWER:**

The target is an **observed outcome** that indicates whether a content page is declining and should be considered for review.

The starter dataset includes the column **is_declining_label**, which represents the target used for supervised learning.

However, this label is derived from historical observations rather than future knowledge. According to the project documentation, related columns such as **trend_direction** and **trend_pct** should not be used as model features because they are directly involved in defining the label.

The model should therefore learn relationships from other observed search, engagement, and content signals rather than simply reproducing the rule that generated the label.

---
## 3. Success metric

*One metric you can defend. What number means 'good'?*

**ANSWER:**

Since the objective is to rank pages according to review priority, the most appropriate evaluation metric is **Precision@K**.

Precision@K measures how many of the top-ranked pages truly require attention. This metric reflects the real workflow of content teams because only a limited number of pages can be reviewed during each cycle.

A higher Precision@K means that more editor time is spent reviewing genuinely important pages rather than low-priority content.

Secondary evaluation metrics such as ROC-AUC or Precision and Recall may also be useful during model development, but Precision@K best represents the practical business decision supported by this project.

## 5. Why ML beats a fixed rule here

*What makes the pattern too messy for an if-statement?*


A fixed rule can only evaluate a small number of conditions at once.

For example, a manually written rule might recommend reviewing a page when:

- CTR is low,
- impressions are decreasing,
- and average position has worsened.

However, real search performance depends on many interacting variables including impressions, clicks, CTR, average position, sessions, engagement rate, freshness, content age, content type, and other observed characteristics.

These relationships are often nonlinear and difficult to express using a few handcrafted rules.

Machine learning can learn complex interactions among these observed signals and generate a ranked recommendation that adapts to patterns present in historical data.

The objective is not to replace human editors but to provide a transparent decision-support system that helps prioritize limited editorial resources.
